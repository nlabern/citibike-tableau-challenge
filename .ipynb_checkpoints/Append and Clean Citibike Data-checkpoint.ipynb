```python
# Append all dataframes to create one main dataframe with same columns
Jul22_to_Jul23_df=pd.concat([Jul22_df, Aug22_df, Sep22_df, Oct22_df, Nov22_df, Dec22_df, 
                             Jan23_df, Feb23_df, Mar23_df, Apr23_df, May23_df, Jun23_df, Jul23_df], ignore_index=True)
Jul22_to_Jul23_df
```

```python
# Check column data types
Jul22_to_Jul23_df.dtypes
```

```python
# Count unique values in each column excluding nulls
unique_counts = Jul22_to_Jul23_df.nunique()
print(unique_counts)
```

```python
# Count null values in all columns
null_counts = Jul22_to_Jul23_df.isnull().sum()
print(null_counts)
```

```python
# Count rows with null start station id & station name
null_count_startstationsid = Jul22_to_Jul23_df['start_station_id'].isnull().sum()
null_count_startstationsname = Jul22_to_Jul23_df['start_station_name'].isnull().sum()
print(null_count_startstationsid)
print(null_count_startstationsname)
```

```python
# Count rows with null end station id & name
null_count_endstations = Jul22_to_Jul23_df['end_station_id'].isnull().sum()
null_count_endstationsname = Jul22_to_Jul23_df['end_station_name'].isnull().sum()
print(null_count_endstations)
print(null_count_endstationsname)
```

```python
# Remove rows that have null station id for start and end stations
df_cleaned_1 = Jul22_to_Jul23_df.dropna(subset=['start_station_id'])
df_cleaned_2 = df_cleaned_1.dropna(subset=['end_station_id'])

# Count null values in all columns
null_counts2 = df_cleaned_2.isnull().sum()
print(null_counts2)
```

```python
# Print unique rideable_type bikes
unique_type = df_cleaned_2['rideable_type'].unique()
print(unique_type)
```

```python
# Print unique member_casual
unique_membercasual = df_cleaned_2['member_casual'].unique()
print(unique_membercasual)
```

```python
# Group by 'start_station_id' and check if there are multiple unique latitudes within each group
stations_with_multiple_latitudes = df_cleaned_2.groupby('start_station_id').filter(lambda group: group['start_lat'].nunique() > 1)

# Get the unique start_station_id' with multiple latitudes
unique_stations = stations_with_multiple_latitudes['start_station_id'].unique()
print(unique_stations)
```

```python
# Group by 'start_station_name' and check if there are multiple unique station names within each group
stations_with_multiple_names = df_cleaned_2.groupby('start_station_id').filter(lambda group: group['start_station_name'].nunique() > 1)

# Get the unique 'start_station_id' with multiple names
unique_stations = stations_with_multiple_names['start_station_id'].unique()
print(unique_stations)
```

```python
# Group by 'end_station_id' and check if there are multiple unique station names within each group
stations_with_multiple_names = df_cleaned_2.groupby('end_station_id').filter(lambda group: group['end_station_name'].nunique() > 1)

# Get the unique 'end_station_id' with multiple names
unique_stations = stations_with_multiple_names['end_station_id'].unique()
print(unique_stations)
```

```python
# Specify the 'end_station_id'
desired_station_id = '5329.08'

# Filter the DataFrame to include only rows with the 'end_station_id'
filtered_df = df_cleaned_2[df_cleaned_2['end_station_id'] == desired_station_id]

# Extract the unique station names associated with the 'end_station_id'
station_names = filtered_df['end_station_name'].unique()
print(station_names)
```

```python
# Specify the 'end_station_id'
desired_station_id2 = '6535.04'

# Filter the DataFrame to include only rows with the 'end_station_id'
filtered2_df = df_cleaned_2[df_cleaned_2['end_station_id'] == desired_station_id2]

# Extract the unique station names associated with the 'end_station_id'
station_names2 = filtered2_df['end_station_name'].unique()
print(station_names2)
```

```python
# Specify the 'end_station_id'
desired_station_id3 = '6708.04'

# Filter the DataFrame to include only rows with the 'end_station_id'
filtered3_df = df_cleaned_2[df_cleaned_2['end_station_id'] == desired_station_id3]

# Extract the unique station names associated with the 'end_station_id'
station_names3 = filtered3_df['end_station_name'].unique()
print(station_names3)
```

```python
# Specify the 'end_station_id'
desired_station_id4 = 'JC115'

# Filter the DataFrame to include only rows with the 'end_station_id'
filtered4_df = df_cleaned_2[df_cleaned_2['end_station_id'] == desired_station_id4]

# Extract the unique station names associated with the 'end_station_id'
station_names4 = filtered4_df['end_station_name'].unique()
print(station_names4)
```

```python
# Create a dictionary to specify the station name changes for station IDs
station_name_changes = {
    '5329.08':'Murray St & West St',
    '6535.04':'W 34 St & Hudson Blvd E',
    '6708.04':'Broadway & W 48 St',
    'JC115': 'Grove St PATH'
}

# Iterate through the dictionary and update the DataFrame
for station_id, new_name in station_name_changes.items():
    df_cleaned_2.loc[df_cleaned_2['end_station_id'] == station_id, 'end_station_name'] = new_name
```

```python
# Group by 'end_station_id' and check if there are multiple unique station names within each group
stations_with_multiple_names = df_cleaned_2.groupby('end_station_id').filter(lambda group: group['end_station_name'].nunique() > 1)

# Get the unique 'end_station_id' with multiple names
unique_stations = stations_with_multiple_names['end_station_id'].unique()
print(unique_stations)
```

```python
# Count unique values in each column excluding nulls
unique_counts = df_cleaned_2.nunique()
print(unique_counts)
```

```python
# Storing file in folder
output_path = Path("Jul22_to_Jul23.csv")
df_cleaned_2.to_csv(output_path, index=None, header=True)
```
